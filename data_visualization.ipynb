```python
#number of goals scored per team

team_goal = data[(data.G > 0) ]
print(team_goal.shape)

```

```python
#goals scored
team_goal = pd.DataFrame(team_goal['TEAM'].value_counts())
print(team_goal)

team_goal.plot.bar(color = 'blue', figsize = (20, 7))
```

```python
plt.figure(figsize=(25,5))
sns.boxplot(x=data['TEAM'],y=data['G'])
plt.savefig('Goals scored per team')
```

```python
#graph to show shots taken and that made it to the goal
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.scatterplot(x=data['SHOTS'], y=data['SOG'])
plt.title ('Shots taken Vs shots on goal')
plt.xlabel ('Shots taken')
plt.ylabel('Shots made it to goal')

```

```python
#grapgh yo show the goals scored compared to shots that made it to goal

plt.subplot(1,2,2)
sns.scatterplot(x=data['SOG'], y=data['G'])
plt.title ('shots on goal vs Goals scored')
plt.xlabel ('shots on goal')
plt.ylabel('Goals scored')
```
